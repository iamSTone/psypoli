## 1) 부동산의 정의

#### 토지와 건물

* 부동산 = 토지 + 건물
* 토지와 건물은 매두 다른 속성을 가진다
    - 건물 : 장기적으로 건물 공급은 완전 탄력적이거나 매우 탄력적이다.
    - 토지 : 토지의 공급은 비탄력적이다. 
        - 국가 전체 혹은 도시 전체 토지 - 완전 비탄력적
        - 주거용, 상업용 등 특정 용도 토지 - 가변적

#### 용도에 따른 구분
* 건물을 포함한 부동산 = 주거용 + 비주거용(=오피스, 판매시설, 산업용, 호텔 등)

#### 수익 창출 여부에 따른 구분
* 자가주택(수익발생x) + 수익용 부동산(income producting property, = 임대주택 + 비주거용 부동산)

## 2) 부동산의 특성

1. 토지의 공급이 고정되어 있다.
    - 토지의 공급량은 가격이 높든, 낮든 잘 변하지 않는다.
    - 주택은 토지와 달리 늘거나 줄 수 있다.
        - 토지와 자본을 결합시켜 생산가능
        - 유지보수(자본투입)을 안하면 노후화 및 멸실한다.
        - 다만 일반 재화에 비해 생산에 긴 시간이 요구된다.
            - 그래서 단기와 장기의 공급 반응이 크게 다르다. 
2. 모든 토지는 서로 다르다.(토지간의 대체가 거의 불가능)
    - 다른 재화에 비해, 토지는 토지간의 이질성이 심히다.
    - 위치 고정성
        - 동일한 2개의 위치란 없다.
        - 토지의 가치는 '첫째도 위치, 둘째도 위치, 셋째도 위치'에 의해 결정된다.
        - 위치의 고정성에서 '독점력'이 생김.
3. 토지는 인간생존에 필수적인 생산요소이다.
    - 형평성(공정성) 
        - 이 때문에 토지소유권 또는 이용권의 공평한 배분이나 그 가격의 적정성이 형평차원에서 중요한 문제로 부각된다.
    - 생산요소로서의 특성 - 자본재에 가까움
        - 토지는 엄밀히 볼 때 저량(stock) 개념인 토지 그 자체보다는 
        - 토지가 제공하는 유량(flow) 개념의 서비스다.
            - 생산재보다는 자본재(공장이나 기계 등과 같은)에 가깝다.
            - 토지도 '토지 서비스의 흐름'을 제공한다.
                - 토지 사용자는 이 서비스 흐름의 대가로 토지 소유자에게 임대료를 지불.
                - 토지 소유자는, 자기자신에게 임대료를 주고 받는다고 생각하는 것이 편리.
    - 주택은 주거 서비스를 제공하는 자본재로 볼 수 있다.
        - 공공성
            - 소득이 높든, 낮든 국민 누구나 인간의 존엄을 지키며 건강한 주거생활을 하는 것이 대부분의 나라의 주택정책의 핵심목표
                - 그런 의미에서 주택은 공공적인 성격을 가짐.
4. 토지와 주택을 포함하는 모든 부동산은 자산이기도 하다.
    - 부동산 자산은 개인이 가진 부의 중요한 구성요소이며 소득의 원천이며, 국부의 상당한 비중을 차지한다.
    - 자산으로서의 부동산 가격은 '매매가격'이다. 
5. 부동산은 금융이 중요한 재화이다.
    - 보통 부동산 구입에는 대출이 필요하다.
    - 공급 측면에서는 부동산 개발 사업 자금은 프로젝트 파이낸싱에 의해 조달된다.

## 3) 부동산 시장의 특성

#### 시장분할(market segmentation)
* 부동산 시장의 본질적 특성 중의 하나는 시장분할(market segmentation)이다.
    - 국가 전체의 주택시장은 존재하지 않는다. 한 도시의 주택시장은 하나의 동질적 시장이 아니라 주택 유형, 규모, 품질 등에 따라 구별되는 상대적으로 규모가 작은 다수의 하부시장 또는 하위시장(submarkets)으로 구성된다.

#### 높은 거래비용
* 부동산의 거래에는 다른 재화에 비해 높은 거래비용이 수반된다.
    - 거래비용의 가장 중요한 항목은 세금과 중개수수료이다.
    - 주택거래비용 = 세금 + 중개수수료 + 주택 탐색비용 + 이사비용(=금전적 비용 + 정신적 비용) 

#### 공실(vacancies)의 존재
* 부동산 시장의 또 하나 특성은 균형상태에서도 공실(vacancies)이 존재한다는 것이다.
    - 신규 부동산 공급에는 많은 비용과 시간이 소요
    - 수요 증가에 대해 공급 시간이 차이가 난다.
* 완충장치(buffer)
    - 건물 임대업자의 입장에서는 수요 증가에 대비하여 여유 공간을 보유하는 것이 바람직.
    - 공실은 곤간에 대한 수요 증가를 큰 비용 들이지 않고 수용할 수 있는 완충장치(buffer)라고 할 수 있다.
* 탐색 가능
    - 공실은 공간 수요자들의 탐색을 돕고 따라서 공간 수요와 공급이 원활하게 연결되도록 하는 기능을 한다.
* 시장 지배력
    - 부동산 시장에서는 수요자와 공급자가 모두 어느 정도의 시장 지배력을 지님
    - 잠재적 구매자와 공급자가 연결되면 거래 가격은 양자 간의 협상에 의해 결정. 
* 초과 공급
    - 부동산은 이질적 재화이며 (이론적인 경제적)균형에서도 공실, 즉 초과 공급이 존재.
* 가격 순응
    - 부동산 시장에서는 수요자나 공급자가 가격순응자가 아닐 수 있다.

#### 가격과 수량 조정이 즉각적으로 이루어지지 않을 수 있다
* 부동산 시장에서는 가격과 수량 조정이 즉각적으로 이루어지지 않을 수 있다.
* 수량 조정
    - 매년 공급되는 신규 공급 규모는 기존 부동산 재고에 비해 매우 작다. -> 부동산 재고는 서서히 증가
    - 일정 시점에서는 부동산 공간 서비스는 기존 건물 재고에서 창출
    - 고층 오피스 건물 등 일부 부동산은 긴규 공급 증가 속도가 더디다.
* 가격 조정
    - 가격 조정은 적정 공실률을 유지하는 범위 내에서 이루어진다.
    - 수요가 증가할 경우
        - 임대료 인상을 통해서 수요증가를 전부 수용할 경우 공실이 지나치게 낮아질 수 있다 -> 임대료를 충분히 인상하지 않는다
    - 수요가 위축될 경우
        - 임대료를 인하하여 임차공간을 늘리기보다는 -> 임대료를 인하하지 않고 적정 공실률을 유지하려고 한다.
    - 상업용 부동산 시장
        - 상업용 부동산 시장에서 가격 조정이 서서히 이루어지는 또 다른 이유는 -> 장기 임대차 계약이 많기 때문이다.
    - 자가주택 시장
        - 주택가격이 하락할 경우 소유주들이 가격하락을 수용하지 않음 -> 가격이 하락하기 보다는 거래량이 감소하는 경향.

#### 주거용 부동산과 비주거용 부동산의 특징
* 공통성
    - 시장 기본 가치(market fundamentals)를 결정하는 생산성 및 지역 경제 변수들은 유사하다.
* 차별성
    - 재정거래 또는 차익거래(arbitrage)
        - 상업용 부동산 시장에 비해 주택시장에서는 재정거래 또는 차익거래가 제한적이다.
        - 그래서 주택가격은 시잔 기본 가치에서 벗어날 가능성이 높다.
    - 전문성
        - 상업용 부동산 시장의 거래는 전문적 투자들이 주도
        - 자가주택 시장은 비전문가인 자가 보유가구들이 주도 -> 심리적 요인의 영향을 많이 받는다. 

#### 개발된 토지와 미개발된 토지
* 개발된 토지 
    - 토지는 한번 개발하고 나면 개발 이전 상태로 되돌리기가 매우 어렵다.
* 미개발 토지
    - 실물옵션(real opton)
        - 미래 시점에 일정 가격(건축비용)을 지불하고 개발된 건물이라는 실물 자산을 구입할 수 있는 권리
    - 즉, 미래에 개발할 수 있는 옵션 가치가 포함된다.

#### 부동산 경기변동
* 부동산 경기변동은 거시경제 변동에서 중요한 변수다.

### (1) 주택에 대한 정부개입

#### (가) 정부개입의 목적

* 부동산은 어느 나라에서든 다른 모든 재화에 비해 정뷰 규제가 많은 재화.
* 주택 시장에 대한 정부개입의 근거
    - 효율성
    - 형평성

#### 정치적 차원
* 자가주택 보유 촉진은 중산층 형성과 사회의 안정성에 기여
* 저소득층에 대한 주거지원은 사회적 결속의 강화를 위해 필수적

#### (나) 주택정책의 수단

<img src="figures/cap2.4.jpg" width=600 />

주택 시장에 대한 정부 개입은 
* 수요 측면 / 공급 측면
* 정책수단 
    - 직접 시장 개입
    - 세제 및 보조금
    - 규제

#### 직접 시장 개입 
* 공공임대주택 직접 생산, 공급
* 주택의 거래나 배분을 통제

#### 수요 측면 + 세제/보조금
* 주택에 대한 재산세, 양도소득세
* 거래관련 세금
* 주택대출자금 이자에 대한 소득 공제
* 주택수당

#### 공급 측면 + 규제/세제

#### 미국의 경우
* 가장 중요한 정책수단은 
    - 자가보유 가구의 주택대출 이자상환액
    - 지방세인 재산세 납부액을 연방 소득세의 과세대상 소득에서 공제하는 것
* 주거주주택(primary residence)에 대해서는 실질적으로 양도소득세가 부과되지 않는다.
* 금융측면
    - 정부기관 또는 정부후원기관(Govemment Sponsored Enterprises: GSEs)의 보증을 통해
        - 장기 고정금리 모기지 공급
* 임차인 지원
    - 공공임대주택의 직접공급 -> (전환) -> 일정조건을 충족하는 저소득층용 주택을 공급하는 민간 건설업체에 대한 세제혜택 부여

#### 유럽의 경우
* 공공임대주택 공급
    - 영국은 기존 임차인들에게 공공임대주택을 할인된 가격에 분양

#### 싱가포르
* 대부분 공공임대주택은 임대용이지만, 싱가포르에서는 국가가 직접 분양. 국민의 89%가 주택을 보유

#### 우리나라
* 주택부문 규제가 다양하고 강력한 나라 중 하나
* 우리나라만의 규제
    - 신규아파트 분양가 규제 및 평형 규제
    - 재건축 관련 다양한 규제
    - 주택거래에 대한 규제
    - 택지공급
        - 주택생산의 핵심요소인 택지공급은 공공부문에 의해 좌우
        - 민간의 역할은 소규모 택지개발에 국한
    - 세제
        - 소수 고가주택에 대해 부과되는 종합부동산세
        - 1가구 다주택 보유에 대한 양도소득세 중과세 

## 2) 주택시장의 필터링 과정과 정책 효과
* (1) 필터링 이론
* (2) 신규주택에 대한 정부지원의 효과
* (3) 필터링과 우리나라 주택정책

### (1) 필터링 이론

* 필터링(filtering) 또는 여과과정은 시간이 경과함에 따라 주택의 질과 주택에 거주하는 소득이 어떻게 변화하는지를 설명하는 이론이다.
* 필터링 과정의 두 가지 기본 과정
    - 첫째. 시간이 흐름에 따라 -> 주택의 질이 감소한다. 
        - 물리적으로 노후화되고, 
        - 기술적으로 낙후되거나 
        - 유행에 뒤처지게 되어
    - 둘째, 주택의 질이 하락 -> 소득이 더 낮은 계층이 더 낮은 가격에 그 주택을 구매한다. 

#### 그림 1-3 설명
* 주택의 '질' 3 단계 - 30, 20, 10
* 소득 '계층' 3 종류 - h(고), m(중), l(저)
* 10년 경과하면 '질'이 10단위 -> 즉 1 단계 하락
    - 주택 보유자는 
        - 집을 보수하거나
        - 이사한다.
* 보수하지 않고 이사하면, 그 집을 다른 계층이 산다.

<img src="figures/cap2.5.jpg" width=600 />

#### 결론

* 이처럼 기존 주택의 질이 저하됨에 따라 각 주택은 더 낮은 소득계층에게 순차적으로 넘어가지만 각 소득계층의 가구는 원래 주거 수준을 회복할 수 있다.
* 이것이 가능한 이유는 가장 질이 낮아 주택재고에서 퇴장하는 집 대신 가장 높은 질의 신규주택이 공급되고 이 주택을 기존의 고소득층이 구입함에 따라 기존의 가장 높은 주택이 다음 소득계층에게 매각되기 때문이다.

### (2) 신규주택에 대한 정부지원의 효과

#### 그림 1-4 설명
* 보조금이 지급되면
* 고소득층은 h점에서 시작
    - 집을 보수 할까?
    - 살까?
* 즉, 고소득층이 원하는 집의 질적 수준이 35로 높아진다.
* 다른 소득 계층들도 집을 옮기기에 고려하는 질적 수준 지점이 5씩 땡겨진다.

#### 결론
* 정부는 고소득층이 구입하는 신축주택에 보조금을 지불하자 -> 필터링 과정 작동 -> 모든 소득 계층의 질적 주거 수준이 각각 5단위씩 높아진다.
* 만일 시간이 경과함에 따라 모든 소득계층의 소득이 상승하면 필터링으로 인한 이득은 더욱 커진다.

### (3) 필터링과 우리나라 주택정책

* 실제로 필터링은 신규주택의 건설 확대를 통해 주택문제를 해결하려는 우리나라 정책의 전제가 되어 왔다.
* 그러나 우리나라 주택시장에서 필터링 과정이 작동하는지 분석한 실증적 연구는 별로 없다.
* 필터링 작동 과정과 그 효과에 대해서는 보다 체계젹인 연구가 필요하다.
